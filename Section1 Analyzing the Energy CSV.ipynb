```r
library(ggplot2)


ggplot(data, aes(x = Wall.Area, y = Heating.Load)) +
  geom_point() +
  labs(title = "Scatter Plot", x = "Wall Area", y = "Heating Load")
```

```r
library(dplyr)
library(ggplot2)

transformed_data <- data %>%
  group_by(Relative.Compactness) %>%
  summarise(mean_heating_load = mean(Heating.Load))

ggplot(transformed_data, aes(x = Relative.Compactness, y = mean_heating_load)) +
  geom_bar(stat = "identity") +
  labs(title = "Bar Plot", x = "Relative Compactness", y = "Mean Heating Load")
```

```r
library(tidyr)
library(ggplot2)


long_data <- data %>%
  gather(key = "area_type", value = "area_value", Wall.Area)

ggplot(long_data, aes(x = area_value, y = Heating.Load, color = area_type)) +
  geom_line() +
  labs(title = "Line Plot", x = "Area", y = "Heating Load")
```
